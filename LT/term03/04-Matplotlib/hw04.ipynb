#### Задача 1.

Постройте график для функции $y = x^2$. Настойчиво предлагаю поиграться с его визуальными настройками. 

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-10, 10.01, 0.01)
plt.plot(x, x ** 2)
plt.show()
```

```python
import plotly.graph_objects as go
import numpy as np

def plot_function():
    # Генерируем значения x от -10 до 10 с шагом 0.1
    x = np.arange(-10, 10, 0.1)

    # Вычисляем значения y = x²
    y = x**2

    # Создаем график с помощью plotly
    fig = go.Figure(data=go.Scatter(x=x, y=y))

    # Настраиваем оси и заголовок графика
    fig.update_layout(
        xaxis_title='X',
        yaxis_title='Y',
        title='Graph of y = x²'
    )

    # Отображаем график
    fig.show()

plot_function()
```

```python
import seaborn as sns
from ipywidgets import interact, Dropdown

def view_palette(palette):
    sns.set_palette(palette)
    sns.palplot(sns.color_palette())

# Список доступных палитр seaborn
palettes = ['deep', 'muted', 'bright', 'pastel', 'dark', 'colorblind']

# Создание выпадающего списка с палитрами
palette_dropdown = Dropdown(options=palettes, description='Palette')

# Обертка функции в interact
interact(view_palette, palette=palette_dropdown)
```
